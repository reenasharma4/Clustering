#Clustering

#1. What is unsupervised learning in the context of machine learning?
  - Unsupervised learning is a type of machine learning where the algorithm learns patterns from data without any labels or guidance. The system tries to group, organize, or find structure in the data on its own. Clustering is one of the most common techniques in unsupervised learning.

#2. How does K-Means clustering algorithm work?
  - K-Means is a method that divides data into "k" groups or clusters. It starts by randomly placing "k" points called centroids. Then, it assigns each data point to the nearest centroid. After all points are assigned, the centroids move to the center of their assigned points. This process repeats until the centroids no longer move much.

#3. Explain the concept of a dendrogram in hierarchical clustering.
  - A dendrogram is a tree-like diagram used in hierarchical clustering. It shows how data points are merged together step by step. At the bottom, each data point starts as its own cluster. As we move up, similar clusters are joined together, and the tree grows. It helps us visualize and decide how many clusters we should have.

#4. What is the main difference between K-Means and Hierarchical Clustering?
  - The main difference is how they form clusters. K-Means creates clusters by moving centroids and requires us to choose the number of clusters beforehand. Hierarchical clustering builds a tree of clusters without needing to choose the number of clusters in advance. You can later cut the tree at any level to get the desired number of clusters.

#5. What are the advantages of DBSCAN over K-Means?
  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise) can find clusters of any shape, while K-Means works best with round-shaped clusters. DBSCAN also does not require the number of clusters to be specified in advance and is better at detecting noise or outlier points that do not belong to any cluster.

#6. When would you use Silhouette Score in clustering?
  - Silhouette Score is used to measure how well the data has been clustered. It tells us if the points within a cluster are similar to each other and different from points in other clusters. A higher score means better-defined clusters. It is helpful to decide the right number of clusters or to compare clustering results.

#7. What are the limitations of Hierarchical Clustering?
  - Hierarchical clustering can be slow and uses a lot of memory when dealing with large datasets. Once a step is done (like merging two clusters), it ca not be undone. Also, it can be sensitive to noise and does not work well if clusters are not clearly separated.

#8. Why is feature scaling important in clustering algorithms like K-Means?
  - K-Means uses distance to group points, so if one feature has much larger values than another, it can unfairly dominate the results. Feature scaling (like normalization or standardization) puts all features on the same scale, so each one contributes equally to the clustering.

#9. How does DBSCAN identify noise points?
  - DBSCAN looks for areas in the data where points are packed closely together (dense regions). Points that do not belong to any dense area and do not have enough neighbors nearby are marked as noise or outliers. These points are not assigned to any cluster.

#10. Define inertia in the context of K-Mean.
   - Inertia is a measure of how spread out the points are within each cluster in K-Means. It calculates the sum of distances between each point and its cluster center. Lower inertia means the points are closer to their cluster centers, which usually indicates better clustering.

#11. What is the elbow method in K-Means clustering?
   - The elbow method helps us choose the best number of clusters for K-Means. We run K-Means with different numbers of clusters and plot the "inertia" (within-cluster sum of distances) for each one. As the number of clusters increases, the inertia decreases. At some point, the decrease becomes very smallâ€”this point looks like an "elbow" in the graph. That is usually the best number of clusters.

#12. Describe the concept of "density" in DBSCAN.
   - In DBSCAN, "density" means how many data points are packed closely together in an area. If an area has enough points within a small distance (called epsilon), it's considered a dense region and forms a cluster. Areas with very few points are considered sparse or noise.

#13. Can hierarchical clustering be used on categorical data?
   - Yes, hierarchical clustering can be used on categorical data, but it requires a suitable way to measure distances or similarities, like the Hamming distance or Jaccard similarity. You ca not use regular distance formulas like Euclidean distance on categorical data directly.

#14. What does a negative Silhouette Score indicate?
   - A negative Silhouette Score means that a point might be assigned to the wrong cluster. It is closer to a different cluster than the one it currently belongs to. This usually shows that the clustering result is not good.

#15. Explain the term "linkage criteria" in hierarchical clustering.
   - Linkage criteria determine how the distance between two clusters is calculated in hierarchical clustering. Some common types are:
     * Single linkage: shortest distance between any two points in the clusters.
     * Complete linkage: longest distance between any two points.
     * Average linkage: average distance between all pairs of points.

These affect how clusters are merged as the algorithm runs.

#16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
  - K-Means assumes that all clusters are about the same size and density and that they are round. If clusters are of different shapes, sizes, or densities, K-Means can group the data incorrectly because it uses a simple distance-based approach.

#17. What are the core parameters in DBSCAN, and how do they influence clustering?
   - DBSCAN has two main parameters:
     * eps (epsilon): The maximum distance between two points to be considered neighbors.
     * min_samples: The minimum number of points needed to form a dense region.

If eps is too small or min_samples is too high, it may label too many points as noise. If eps is too large, different clusters may get merged incorrectly.

#18. How does K-Means++ improve upon standard K-Means initialization?
   - K-Means++ chooses the initial centroids more carefully by spreading them out. Instead of picking them randomly, it picks the first one at random, and each next centroid is chosen based on how far it is from already chosen centroids. This often leads to better and more stable clustering results.

#19. What is agglomerative clustering?
   - Agglomerative clustering is a type of hierarchical clustering that builds clusters from the bottom up. It starts with each point as its own cluster and then merges the closest clusters step by step until everything is combined into one big cluster or until a stopping point is reached.

#20. What makes Silhouette Score a better metric than just inertia for model evaluation?
   - Inertia only tells how tightly the points are grouped within clusters but does not consider how well-separated the clusters are. Silhouette Score looks at both: how close a point is to its own cluster and how far it is from other clusters. This gives a more complete picture of how good the clustering is.



```python
#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate data
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            color='red', marker='x', s=200, label='Centroids')
plt.title("K-Means Clustering on make_blobs Data")
plt.legend()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
from sklearn.metrics import silhouette_score
import numpy as np

# Generate synthetic data
X, _ = make_blobs(n_samples=400, centers=4, cluster_std=0.6, random_state=0)

# Evaluate silhouette scores
scores = []
k_values = range(2, 6)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Plot silhouette scores
plt.bar(k_values, scores, color='skyblue')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
#30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# Load data
iris = load_iris()
X = iris.data

# Compute linkage matrix
linkage_matrix = linkage(X, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix, labels=iris.target, leaf_rotation=90)
plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate data
X, _ = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

# Fit K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.predict(X)

# Create meshgrid for decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.contourf(xx, yy, Z, alpha=0.3, cmap='Accent')
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Accent', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='x', s=200, label='Centroids')
plt.title("K-Means with Decision Boundaries")
plt.legend()
plt.show()
```

```python
#34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Load and scale data
data = load_breast_cancer()
X = StandardScaler().fit_transform(data.data)

# Inertia values for k = 2 to 6
inertias = []
k_values = range(2, 7)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot
plt.plot(k_values, inertias, marker='o')
plt.title("Inertia vs Number of Clusters (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid()
plt.show()
```
